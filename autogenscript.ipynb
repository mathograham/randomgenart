```python
#using numpy library for random number generator
import numpy as np

def generateRandomNumber(lowIn, highIn):
    rng = np.random.default_rng()
    ranNumber = rng.integers(low = lowIn, high = highIn)
    return int(ranNumber)

#if lowIn=0, then characteristic is optional and may not be included at all
#for increased rarity, make highIn well outside range of dictionary values
#the higher the range, the more rare the characteristics listed will be

#background: base layer; must be added before everything else


bkgrd = generateRandomNumber(7,10)

print(bkgrd)

def createBkgrdDict():
    BkgrdDict = {
        7: 'light shadow',
        8: 'grey shadow',
        9: 'black shadow',
    }
    
    return BkgrdDict

#egg: layer must be added before hat

egg = generateRandomNumber(1,7)

print(egg)

def createEggDict():
    EggDict = {
        1: 'golden',
        2: 'sulfur',
        3: 'century',
        4: 'farm-fresh',
        5: 'quail',
        6: 'robin',
    }
    return EggDict
#background accent: can be added at any layer after background

bkgdAcc = generateRandomNumber(0,15)

print(bkgdAcc)

def createBkgdAccDict():
    bkgdAccDict = {
        1: 'demon tail',
        2: 'cherub wings',
        3: 'blue morpho wings',
        4: 'monarch wings',
        5: 'bat wings',
        6: 'reptile tail',
        7: 'lemur tail',
    }
    return bkgdAccDict

#top accents: must come after after, preferably before bottom acc. Must come
#before hat

topAcc = generateRandomNumber(0,15)

print(topAcc)

def createTopAccDict():
    topAccDict = {
        1: 'blushing',
        2: 'bandage',
        3: 'cherry cheeks',
        4: 'model mole',
        5: 'clown nose',
        6: 'fully cracked',
        7: 'cracked',
        8: 'chipped',
    }
    return toppAccDict

#bottom accents: comes after top accent.

botAcc = generateRandomNumber(0,25)

print(botAcc)

def createBotAccDict():
    botAccDict = {
        1: 'gold chain',
        2: 'red banana',
        3: 'hot pink banana',
        4: 'black banana',
        5: 'agent orange banana',
        6: 'yellow banana',
        7: 'green machine banana',
        8: 'blue ruin banana',
        9: 'purple banana',
        10: 'green machine fanny pack',
        11: 'hot pink fanny pack',
        12: 'blue ruin fanny pack',
        13: 'agent orange fanny pack',
    }
    return botAccDict


#feet: can come before or after egg

feet = generateRandomNumber(0,31)

print(feet)

def createFeetDict():
    FeetDict = {
        1: 'lavender',
        2: 'red',
        3: 'red hightop',
        4: 'royal',
        5: 'crow',
        6: 'goose',
        7: 'hawk',
        8: 'carrion',
        9: 'eagle',
        10: 'chicken',
        11: 'green revolution hightop',
        12: 'blue ruin',
        13: 'blue ruin hightop',
        14: 'green revolution',
        15: 'hot pink',
        16: 'lavender hightop',
        17: 'green machine hightop',
        18: 'green machine',
        19: 'yellow hightop',
        20: 'yellow',
        21: 'agent orange hightop',
        22: 'agent orange',
        23: 'black hightop',
        24: 'black',
        25: 'all black hightop',
        26: 'green machine skates',
        27: 'blue ruin skates',
        28: 'white skates',
        29: 'hot pick hightop',
    }
    return FeetDict

#hat: hat is last layer

hat = generateRandomNumber(0,30)

print(hat)

def createHatDict():
    HatDict = {
        1: 'black beanie',
        2: 'hot pink beanie',
        3: 'rust beanie',
        4: 'agent orange beanie',
        5: 'green revolution beanie',
        6: 'yellow beanie',
        7: 'green machine beanie',
        8: 'blue ruin beanie',
        9: 'royal beanie',
        10: 'red beanie',
        11: 'rainbow beanie',
        12: 'striped beanie',
        13: 'rust headband',
        14: 'green revolution headband',
        15: 'royal headband',
        16: 'racing headband',
        17: 'red headband',
        18: 'hot pink headband',
        19: 'blue ruin headband',
        20: 'green machine headband',
        21: 'yellow headband',
        22: 'agent orange headband',
        23: 'black headband',
        24: 'headphones',
        25: 'demon stubs',
        26: 'cherub halo',
    }
    return HatDict

#generate token to identify combination that generated nft
token_id = int(str(bkgrd) + str(egg) + str(bkgdAcc).rjust(2, '0') + str(topAcc).rjust(2, '0') + str(botAcc).rjust(2, '0') + str(feet).rjust(2, '0') + str(hat).rjust(2, '0'))
print(token_id)
```

```python
from PIL import Image

# open pngs (background, base, feet, hats) in respective folders

base = Image.open("images/" "egg-background/" + str(bkgrd) + ".png")

eggskn = Image.open("images/" + "egg-base/" + str(egg) + ".png")

if 0< bkgdAcc < 8:
    bkgdaccs = Image.open("images/" + "bkgd-acc/" + str(bkgdAcc) + ".png")

if 0< topAcc< 9:
    topAccs = Image.open("images/" + "top-acc/" + str(topAcc) + ".png")
    
if 0 < botAcc < 14:
    botAccs = Image.open("images/" + "bottom-acc/" + str(botAcc) + ".png")

if 0< feet <30:
    feets = Image.open("images/" + "feet/" + str(feet) + ".png")
    
if 0 < hat < 27:
    headwear = Image.open("images/" + "hats/" + str(hat) + ".png")

print(feet)
print(hat)
# paste/merge required pngs, as layers on base

base.paste(eggskn, (0,0), eggskn)

if 0< feet <30:
    base.paste(feets, (0,0), feets)

if 0< bkgdAcc < 8:
    base.paste(bkgdaccs, (0,0), bkgdaccs)


if 0< topAcc< 9:
    base.paste(topAccs, (0,0), topAccs)

if 0 < botAcc < 14:
    base.paste(botAccs, (0,0), botAccs)
    
    
if 0 < hat < 27:
    base.paste(headwear, (0,0), headwear)

display(base)

base.save('images/a_egg_generated/a_egg' + str(token_id) + '.png', "PNG")

#notes: nfts will likely be on white backgrounds. Some accents are not viewable
#on white backgrounds right now. Change color slighlty so can be seen.
#or outline
#crack looks messy on some of them. Change shape. Looks bad on quail.
#quail looks messy in general. Maybe change quail to something else. Maybe dark brown.
#add goosalagoopagoop tail
```
